**Pikling & Unpikling :**

```
# it is unuseful for untrusted data.
import pickle as pkl

class Student:
    def __init__(self,name,roll,add):
        self.name=name
        self.roll=roll
        self.address=add

    def disp(self):
        print(f'name : {self.name}, roll : {self.roll}, Address : {self.address}')

with open('/content/student.dat',mode='wb') as f:
    stu1=Student('mahfuz', '18esd083', 'dinajpur')
    stu2=Student('piyal', '18esd023', 'kisorganj')
    pkl.dump(stu1, f)
    pkl.dump(stu2, f)
    print('succeed')

with open('/content/student.dat', mode='rb') as f:
    while True:
        try:
            obj=pkl.load(f)
            obj.disp()
        except EOFError:
            break
```
