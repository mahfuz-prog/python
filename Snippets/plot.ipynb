```python
ages_x=[10,20,30,40,50,60]
dev_y=np.sin(ages_x)
py_dev_y=np.cos(ages_x)

#plt.xkcd() it draw the chart like hand written.... comics

plt.plot(ages_x, py_dev_y, color='b', linewidth=3, label='python')
plt.plot(ages_x, dev_y, color='k', linestyle='--', label='All dev')

plt.xlabel('ages')
plt.ylabel('salary')

plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()


#plt.savefig('name.png')
```

```python
eg=np.tan(ages_x)

ages_indexs = np.arange(len(ages_x)) #for side align bars
width=.25

plt.bar(ages_indexs - width, py_dev_y, width = width, linewidth=3, label='python')
plt.bar(ages_indexs, dev_y, width = width, linestyle='--', label='All dev')
plt.bar(ages_indexs + width, eg, width = width, label='exp')

plt.xticks(ticks=ages_indexs, labels=ages_x)  # for fix the ages details in bottom

plt.xlabel('ages')
plt.ylabel('salary')

plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
test_df = df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/pandas demo/survey_results_public.csv', usecols=["LanguageWorkedWith"])

test_df.dropna(inplace=True)
test_df['LanguageWorkedWith'] = test_df['LanguageWorkedWith'].str.split(';')

from collections import Counter

language_counter = Counter()
for _ in df['LanguageWorkedWith']:
    language_counter.update(_)

language_counter = language_counter.most_common(15) # list of tuples
languages, popularity = list(zip(*language_counter))

languages = list(languages)       #
popularity = list(popularity)     # for display the chart in reverce
languages.reverse()               #
popularity.reverse()              #

plt.barh(languages, popularity)

plt.title('Most popular languages')
plt.xlabel('popularity')
plt.ylabel('language')

plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
plt.style.use('classic')

test_df = df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/pandas demo/survey_results_public.csv', usecols=["LanguageWorkedWith"])

test_df.dropna(inplace=True)
test_df['LanguageWorkedWith'] = test_df['LanguageWorkedWith'].str.split(';')

from collections import Counter
language_counter = Counter()
for _ in df['LanguageWorkedWith']:
    language_counter.update(_)

language_counter = language_counter.most_common(5) 
languages, popularity = list(zip(*language_counter))

languages = list(languages)       
popularity = list(popularity) 
explode=[0,0,0,0.1,0]   # for explode item

plt.pie(popularity, labels=languages, explode=explode, shadow=True, autopct='%1.1f%%',
        wedgeprops={'edgecolor': 'black'})

plt.title('Most popular languages')
plt.tight_layout()
plt.show()
```
