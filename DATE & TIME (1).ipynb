**TIME MODULE :**



```
from time import time,ctime,localtime

epoch=time() #time() return current date in S.
print(epoch)

print(ctime())
print(ctime(epoch))
#if parameter pass it will show date accoding perameter
#if no pera pass it will show current date

obj=localtime()
print(obj) #it return all attribute of date and time
print(obj.tm_mon)

```

```
#number of days in a month

class Solution(object):
    def numberOfDays(self, y, m):
        if y >=1700 and y <=1917:

        # julian calendar 
        
            leap = 0
            if y% 4 == 0:
                leap = 1
            if m==2:
                return 28 + leap
            list = [1,3,5,7,8,10,12]
            if m in list:
                return 31
            return 30

        # Gregorian calendar

        else:
            leap = 0
            if y% 400 == 0:
                leap = 1
            elif y% 4 == 0 and y % 100 !=0:
                leap = 1
            if m==2:
                return 28 + leap
            list = [1,3,5,7,8,10,12]
            if m in list:
                return 31
            return 30
ob1 = Solution() 
ob1.numberOfDays(2019, 2)
```
